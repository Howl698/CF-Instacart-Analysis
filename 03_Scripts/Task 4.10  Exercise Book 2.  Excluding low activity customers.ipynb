## 2.  Create an exclusion flag

```python
df_cust.loc[df_cust['order_number'] < 5, 'activity_flag'] = 'low activity'
```

```python
df_cust.loc[df_cust['order_number'] >= 5, 'activity_flag'] = 'active'
```

  
## 3. Exclude low activity customers from the dataframe

```python
# selecting rows based on 'active' condition

name = ['active']

df_cust2 = df_cust.loc[df_cust['activity_flag'].isin(name)]
```
