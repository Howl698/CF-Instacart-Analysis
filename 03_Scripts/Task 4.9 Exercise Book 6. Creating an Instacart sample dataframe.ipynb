## 2. Create a sample set

### Due to the size of the merged dataframe (32404859 rows x 34 columns), a random sample of the data frame will be used for  analysis and visualizations.

```python
# Creating a pattern of random numbers (in this case, seed “4”). Creating a random sample using a 70/30 split (from NumPy, rows assigned less than 0.7 in one sample, greater in the other)

np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe 'big'

big = ords_prods_merge[dev]
```

```python
big.describe()
```

```python
# export big sample dataframe

big.to_pickle(os.path.join(path,'02_Data','Prepared_Data', 'big_cop_sample.pkl'))
```

```python
# Store 30% of the sample in the dataframe 'small'

small = ords_prods_merge[~dev]
```

```python
small.describe()
```
